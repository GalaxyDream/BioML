```python
import pandas as pd
from pandas_summary import DataFrameSummary
import numpy as np
pd.set_option('display.max_columns', 100)
```

## Load dataset

#### Read headers of csv file from the gene file

```python
with open('../data/Genes.txt', 'r') as f:
    genes = f.readlines()
```

```python
headers = [x.split()[0] for x in genes]
```

```python
headers.insert(0, 'label')
```

```python
len(headers)
```

#### Read data from csv file

```python
data = pd.read_csv('../data/HW3Data.csv', names = headers)
```

```python
len(data.columns)
```

```python
data.head()
```

```python
DataFrameSummary(data).summary()
```

```python
# X includes all other columns except "label"; y is the value of column "label"
X = data[data.columns[1:]].values
y = data['label'].values
```

## Train test split (20% hould out)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, 
                     test_size=0.20,
                     stratify=y,
                     random_state=1)
```

```python
print("X_train shape: ", X_train.shape)
print("Y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("Y_test shape: ", y_test.shape)
```

## Data standardization

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
```

```python
x_train = scaler.transform(X_train)
```

## Recursive Feature Elimination with Cross-Validation (Logistic Regression as Base Classifier)

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LogisticRegression
```

```python
lr_clf = LogisticRegression(penalty = 'l1')
```

```python
rfe_lr = RFECV(estimator=lr_clf, step=1, cv=3,
              scoring='accuracy')
```

```python
rfe_lr.fit(x_train, y_train)
```

```python
print("Optimal number of features : %d" % rfe_lr.n_features_)
```

## Plot curve

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score (nb of correct classifications)")
plt.plot(range(1, len(rfe_lr.grid_scores_) + 1), rfe_lr.grid_scores_)
plt.show()
```

## Use hold out for test

#### Data standardization for test data

```python
x_test = scaler.transform(X_test)
```

#### Test with selected features

```python
from sklearn.metrics import precision_recall_fscore_support as prf
from sklearn.metrics import accuracy_score as acc
from sklearn.metrics import auc, roc_curve
```

```python
y_pred = rfe_lr.predict(x_test)
y_true = y_test
y_pred_prob = rfe_lr.predict_proba(x_test)
fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob[:,1])
```

```python
precision, recall, f_score, _ = prf(y_true = y_true, y_pred = y_pred, average = 'macro')
accuracy = acc(y_true = y_true, y_pred = y_pred)
roc_auc = auc(fpr, tpr)
print('precision: ', precision, 'recall: ', recall, 'f-score: ', f_score, 'accuracy: ', accuracy)
print('roc_auc: ', roc_auc)
```
